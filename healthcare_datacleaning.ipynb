```python
df = df.rename(columns={
    "Month": "Month",
    "Average Licensed Bed  Occupancy Rate": "OccupancyRate",
    "Average Licensed Bed Occupancy Rate": "OccupancyRate",
    "Unassisted Fall Rate per 1,000 Patient Days": "FallRatePer1000PatientDays",
    "Staff Responsiveness Domain Top Box Score": "StaffResponsivenessScore",
    "Unassisted Fall %": "FallRatePercent",
    "Staff Responsiveness Domain Top Box Score %": "StaffResponsivenessPercent",
    "Staff Responsiveness Top Box Score Benchmark": "StaffResponsivenessBenchmark"
})

print(df.columns.tolist())
```

```python
df["Month"] = pd.to_datetime(df["Month"])
```

```python
df["Year"] = df["Month"].dt.year
df["MonthName"] = df["Month"].dt.month_name()
df["MonthNumber"] = df["Month"].dt.month
df["Quarter"] = "Q" + df["Month"].dt.quarter.astype(str)
df["YearMonth"] = df["Month"].dt.to_period("M").astype(str)

df.head()
```

```python
df["StaffResponsivenessBenchmarkStatus"] = np.where(         #tells whether staff responsiveness is above or below benchmark.
    df["StaffResponsivenessScore"] >= df["StaffResponsivenessBenchmark"],
    "At or Above Benchmark",
    "Below Benchmark"
)
```

```python
df["OccupancyCategory"] = pd.cut(           # occupancy category
    df["OccupancyRate"],
    bins=[0, 0.90, 0.95, 1.00, np.inf],
    labels=["Low Occupancy", "Moderate Occupancy", "High Occupancy", "Very High Occupancy"]
)
```
